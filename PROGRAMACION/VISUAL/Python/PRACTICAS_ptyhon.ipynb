Entre 10000€ y 20000€	15%
Entre 20000€ y 35000€	20%
Entre 35000€ y 60000€	30%

EJERCICIO 6


```python
n = int(input("Introduce la altura del árbol (entero positivo): "))

# Ancho máximo del árbol (base impar para centrar bien)
ancho = 2 * n - 1

# Parte superior del árbol (simétrica)
for i in range(n):
    espacios = " " * ((ancho - (2 * i + 1)) // 2)
    estrellas = "*" * (2 * i + 1)
    print(espacios + estrellas)

# Tronco simple
tronco_espacios = " " * ((ancho - 3) // 2)
print(tronco_espacios + "|||")
print(tronco_espacios + "|||")
```
